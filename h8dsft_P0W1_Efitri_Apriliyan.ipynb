# Cleaning Data

Data yang akan kita gunakan selanjutnya adalah data date, confirmed, death, recovered  dan active. Di atas terlihat jika terdapat Missing Values pada data **recovered** dan **active**, maka kita akan mengisi data yang kosong tersebut dengan 0.

```python
#mengisi data yang kosong dengan angka 0
df['recovered'].fillna(0, inplace=True)
df['active'].fillna(0, inplace=True)
```

```python
#mengecek data yang sudah di cleaning pada kolom recovered dan active. NaN berubah menjadi 0
df.head(10)
```

# Rename Columns

```python
#mengganti nama kolom pada data yang akan digunakan selanjutnya

table1= df.rename(columns={'date':'tanggal','country_region':'country','confirmed':'terkonfirmasi','deaths':'meninggal','recovered':'sembuh'})
table1
```

# Drop Columns

```python
#menghapus beberapa kolom yang tidak digunakan yaitu province, latitude, longitude, location geom, fips, admin2 dan combined key.

table2 = table1.drop(columns=['province_state', 'country','latitude','longitude','location_geom','fips','admin2','combined_key'])
table2
```

# Sort Column

```python
#Mensortir kolom date agar tanggal berurutan.

table2.sort_values(by='tanggal')
```

# Query

```python
# Menggunakan query lebih dari (>) untuk melihat pada tanggal berapa angka terkonfirmasi melebihi 10.000
table2.query("terkonfirmasi > 10000 ").sort_values(by='tanggal')
```

```python
# Menggunakan query lebih dari sama dengan (>=) untuk melihat pada tanggal berapa ditemukan pertama kali kasus meninggal
table2.query("meninggal >= 1").sort_values(by='tanggal')
```

```python
# Menggunakan query lebih dari sama dengan (>=) untuk melihat pada tanggal berapa pertama kali kasus sembuh
table2.query("sembuh >= 1").sort_values(by='tanggal')
```

# GroupBy

```python
# Menggunakan groupby untuk melihat tabel minimum pada kasus meninggal
table2_min = table2.groupby('tanggal')['meninggal'].min().sort_values()
table2_min
```

```python
# Menggunakan groupby untuk melihat tabel maksimum pada kasus meninggal
table2_max = table2.groupby('tanggal')['meninggal'].max().sort_values()
table2_max
```

```python
# Menggunakan groupby untuk melihat tabel minimum pada kasus sembuh
table2_min = table2.groupby('tanggal')['sembuh'].min().sort_values()
table2_min
```

```python
# Menggunakan groupby untuk melihat tabel minimum pada kasus sembuh
table2_max = table2.groupby('tanggal')['meninggal'].max().sort_values()
table2_max
```
